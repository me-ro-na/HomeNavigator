```python
import time
import os
import json
from datetime import datetime, timedelta
```

```python
current_dir = os.getcwd()
# JSON 파일 상대 경로
current_date = datetime.now().strftime("%y%m%d")
relative_path = f'../static/data/apartment_{current_date}.json'


# 절대 경로 생성
file_path = os.path.join(current_dir, relative_path)


# Load local JSON file
apartment_file_path = file_path

limit = 1
limit = 'all'
with open(apartment_file_path, 'r', encoding='utf-8') as file:
    apartment_data = json.load(file)
if limit == 'all':
    queries = [apartment['address'] for apartment in apartment_data]
    apartment_name_list = [apartment['name'] for apartment in apartment_data]
else:
    limit = int(limit)
    queries = [apartment['address'] for apartment in apartment_data[:limit]]
    apartment_name_list = [apartment['name'] for apartment in apartment_data[:limit]]


destination_list = ["강남역 2호선", "여의도역 9호선", "판교역 신분당선"]
```

```python
# 한국의 공휴일 목록 (예시: 추가 공휴일을 포함하여 업데이트 필요)
holidays = [
    datetime(2024, 1, 1).date(),  # 신정
    datetime(2024, 2, 10).date(),  # 설날
    datetime(2024, 2, 11).date(),  # 설날
    datetime(2024, 2, 12).date(),  # 설날 연휴
    datetime(2024, 3, 1).date(),  # 삼일절
    datetime(2024, 5, 5).date(),  # 어린이날
    datetime(2024, 5, 15).date(),  # 석가탄신일
    datetime(2024, 6, 6).date(),  # 현충일
    datetime(2024, 8, 15).date(),  # 광복절
    datetime(2024, 9, 16).date(),  # 추석
    datetime(2024, 9, 17).date(),  # 추석
    datetime(2024, 9, 18).date(),  # 추석 연휴
    datetime(2024, 10, 3).date(),  # 개천절
    datetime(2024, 10, 9).date(),  # 한글날
    datetime(2024, 12, 25).date()  # 성탄절
]

# 주말 또는 공휴일 확인
def is_weekend_or_holiday(date):
    return date.weekday() >= 5 or date in holidays
```

```python
# Chrome WebDriver 경로 설정
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options

# WebDriver 상대 경로 설정
driver_path = '../static/lib/chromedriver_h1.exe'
# 절대 경로로 변환
abs_driver_path = os.path.abspath(driver_path)

service = Service(executable_path=abs_driver_path)

options = Options()
options.add_argument("--window-size=100,100")

# Service 객체 생성
# service = Service(abs_driver_path)

# WebDriver 객체 생성
driver = webdriver.Chrome(service=service, options=options)

# 네이버 지도 웹사이트 열기
driver.get("https://map.naver.com/")

#  125.0.6422.14
```

```python
# 성공 및 실패 리스트 초기화
success_list = []
error_list = []

# 페이지 로드 대기
time.sleep(5)  # 페이지가 완전히 로드될 때까지 기다림 (적절한 시간으로 조정 필요)

for index, query in enumerate(queries):
    for destination in destination_list:
        try:
            try:
                button = driver.find_elements(By.CLASS_NAME, "btn_navbar")[1]
                button.click()
                print("버튼 클릭 성공")
            except Exception as e:
                print(f"버튼 클릭 실패: {e}")
                error_list.append({"query": query, "destination": destination, "error": "버튼 클릭 실패"})
                continue

            # 페이지 이동 대기
            time.sleep(5)

            # 출발지 입력
            try:
                start_input = driver.find_elements(By.CLASS_NAME, "search_input_box_wrap")[0].find_element(By.CLASS_NAME, "input_search")
                start_input.send_keys(query)
                start_input.send_keys(Keys.RETURN)
                print("출발지 입력 성공")
            except Exception as e:
                print(f"출발지 입력 실패: {e}")
                error_list.append({"query": query, "destination": destination, "error": "출발지 입력 실패"})
                continue

            # 페이지 이동 및 검색 결과 대기
            time.sleep(5)

            # "btn_departure directions_box start" 클래스 버튼 클릭
            try:
                departure_button = driver.find_element(By.CLASS_NAME, "btn_departure.directions_box.start")
                departure_button.click()
                print("출발 버튼 클릭 성공")
            except Exception as e:
                print(f"출발 버튼 클릭 실패: {e}")
                error_list.append({"query": query, "destination": destination, "error": "출발 버튼 클릭 실패"})
                continue

            time.sleep(5)
            # 도착지 입력
            try:
                goal_input = driver.find_elements(By.CLASS_NAME, "search_input_box_wrap")[1].find_element(By.CLASS_NAME, "input_search")
                goal_input.send_keys(destination)
                goal_input.send_keys(Keys.RETURN)
                print("도착지 입력 성공")
            except Exception as e:
                print(f"도착지 입력 실패: {e}")
                error_list.append({"query": query, "destination": destination, "error": "도착지 입력 실패"})
                continue

            # 페이지 이동 및 검색 결과 대기
            time.sleep(5)

            # "btn_departure directions_box goal" 클래스 버튼 클릭
            try:
                goal_button = driver.find_element(By.CLASS_NAME, "btn_departure.directions_box.goal")
                goal_button.click()
                print("도착 버튼 클릭 성공")
            except Exception as e:
                print(f"도착 버튼 클릭 실패: {e}")
                error_list.append({"query": query, "destination": destination, "error": "도착 버튼 클릭 실패"})
                continue

            time.sleep(5)
            # 시간 옵션 버튼 클릭
            try:
                time_option_button = driver.find_elements(By.CLASS_NAME, "btn_time_option.btn_option")[0]
                time_option_button.click()
                print("시간 옵션 버튼 클릭 성공")
            except Exception as e:
                print(f"시간 옵션 버튼 클릭 실패: {e}")
                error_list.append({"query": query, "destination": destination, "error": "시간 옵션 버튼 클릭 실패"})
                continue

            # 페이지 이동 및 시간 설정 대기
            time.sleep(5)

            # 현재 날짜 확인
            now = datetime.now().date()
            # 주말 또는 공휴일일 경우 평일이 될 때까지 이전 버튼 클릭
            while is_weekend_or_holiday(now):
                try:
                    prev_button = driver.find_elements(By.CLASS_NAME, "timeset_btn_prev")[0]
                    prev_button.click()
                    print("이전 버튼 클릭 성공")
                    # 날짜를 이전으로 설정 (예: 하루 전으로 이동)
                    now = now - timedelta(days=1)
                    time.sleep(1)  # 클릭 후 잠시 대기
                except Exception as e:
                    print(f"이전 버튼 클릭 실패: {e}")
                    error_list.append({"query": query, "destination": destination, "error": "이전 버튼 클릭 실패"})
                    break

            # 시간 설정 대기
            time.sleep(2)

            # 시간을 7시로 설정
            try:
                hour_button = driver.find_elements(By.CLASS_NAME, "timeset_option_hour")[0].find_element(By.CLASS_NAME, "timeset_btn_data")
                while hour_button.text != "07시":
                    prev_hour_button = driver.find_elements(By.CLASS_NAME, "timeset_option_hour")[0].find_element(By.CLASS_NAME, "timeset_btn_prev")
                    prev_hour_button.click()
                    time.sleep(1)  # 각 클릭 후 대기
                    hour_button = driver.find_elements(By.CLASS_NAME, "timeset_option_hour")[0].find_element(By.CLASS_NAME, "timeset_btn_data")
                print("시간 설정 성공")
            except Exception as e:
                print(f"시간 설정 실패: {e}")
                error_list.append({"query": query, "destination": destination, "error": "시간 설정 실패"})
                continue

            time.sleep(5)
            # 현재 분 확인 후 10분 빼기
            try:
                minute_button = driver.find_elements(By.CLASS_NAME, "timeset_option_minute")[0].find_element(By.CLASS_NAME, "timeset_btn_data")
                current_minute = int(minute_button.text.replace("분", ""))
                target_minute = current_minute % 10  # 현재 분의 일의 자리수를 가져옴
                target_minute_text = f"0{target_minute}분"
                while minute_button.text != target_minute_text:
                    prev_minute_button = driver.find_elements(By.CLASS_NAME, "timeset_option_minute")[0].find_element(By.CLASS_NAME, "timeset_btn_prev")
                    prev_minute_button.click()
                    time.sleep(1)  # 각 클릭 후 대기
                    minute_button = driver.find_elements(By.CLASS_NAME, "timeset_option_minute")[0].find_element(By.CLASS_NAME, "timeset_btn_data")
                print("분 설정 성공")
            except Exception as e:
                print(f"분 설정 실패: {e}")
                error_list.append({"query": query, "destination": destination, "error": "분 설정 실패"})
                continue

            # 데이터 추출 시도
            try:
                ul_element = driver.find_element(By.ID, "tab_pubtrans_directions").find_element(By.TAG_NAME, "ul")
                li_elements = ul_element.find_elements(By.TAG_NAME, "li")
                
                # 모든 <li> 요소의 텍스트를 한꺼번에 추출
                results = [li.text for li in li_elements[:6]]
                success_list.append({"query": query, "apartment_name": apartment_name_list[index], "destination": destination, "result": results})
            except Exception as e:
                print(f"데이터 추출 실패: {e}")
                error_list.append({"query": query, "destination": destination, "error": "데이터 추출 실패"})

        except Exception as e:
            print(f"알 수 없는 오류: {e}")
            error_list.append({"query": query, "destination": destination, "error": "알 수 없는 오류"})

# 성공 및 실패 리스트를 JSON 파일로 저장
with open('success.json', 'w', encoding='utf-8') as f:
    json.dump(success_list, f, ensure_ascii=False, indent=4)

with open('error.json', 'w', encoding='utf-8') as f:
    json.dump(error_list, f, ensure_ascii=False, indent=4)

# 성공 및 실패 리스트 출력
print("성공 리스트:", json.dumps(success_list, ensure_ascii=False, indent=4))
print("실패 리스트:", json.dumps(error_list, ensure_ascii=False, indent=4))

# WebDriver 종료
driver.quit()
```
